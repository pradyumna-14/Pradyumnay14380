```python
df.replace('?', np.nan, inplace=True)
df
```

```python
df.dropna(inplace=True)
df
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Select numerical columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
num_cols
```

```python
# Standard Scaling
scaler_std = StandardScaler()
df_std = df.copy()
df_std[num_cols] = scaler_std.fit_transform(df_std[num_cols])
df_std
```

```python
# Min-Max Scaling
scaler_minmax = MinMaxScaler()
df_minmax = df.copy()
df_minmax[num_cols] = scaler_minmax.fit_transform(df_minmax[num_cols])
df_minmax
```

```python
from sklearn.preprocessing import LabelEncoder

df_encoded = df.copy()

# One-Hot Encoding for categorical columns with <= 5 unique values
for col in df_encoded.select_dtypes(include='object').columns:
    if df_encoded[col].nunique() <= 5:
        df_encoded = pd.get_dummies(df_encoded, columns=[col], drop_first=True)
    else:
        le = LabelEncoder()
        df_encoded[col] = le.fit_transform(df_encoded[col])
        print(df_encoded)
```

```python
df_fe = df_encoded.copy()

# New Feature 1: Age buckets
df_fe['age_group'] = pd.cut(df_fe['age'], bins=[0, 25, 45, 65, 100], labels=['Young', 'Adult', 'Senior', 'Elder'])

# New Feature 2: Hours worked per week bucket
df_fe['hours_per_week'] = pd.cut(df_fe['hours_per_week'], bins=[0, 20, 40, 60, 100], labels=['Low', 'Normal', 'High', 'Extreme'])
df_fe
```

```python
import scipy.stats as stats

# Skew check
print(df_fe['capital_gain'].skew())

# Log transformation
df_fe['log_capital_gain'] = np.log1p(df_fe['capital_gain'])

# Compare distributions
sns.histplot(df['capital_gain'], kde=True).set(title='Original Capital Gain')
plt.show()
sns.histplot(df_fe['log_capital_gain'], kde=True).set(title='Log Transformed Capital Gain')
plt.show()
```

```python
from sklearn.ensemble import IsolationForest

iso = IsolationForest(contamination=0.02, random_state=42)
outliers = iso.fit_predict(df_fe.select_dtypes(include=['int64', 'float64']))
df_clean = df_fe[outliers == 1]

print(f"Removed {df_fe.shape[0] - df_clean.shape[0]} outliers")
```
