```python
import pandas as pd
# Load dataset
```

```python
salesd=pd.read_csv(r"C:\Users\91808\OneDrive\Pictures\Desktop\sales_data_with_discounts.csv")
salesd
```

```python
salesd.columns
```

```python
#identify numerical columns
numerical_cols = salesd.select_dtypes(include=['number']).columns
```

```python
# Compute basic statistics
stats_summary = salesd[numerical_cols].agg(['mean', 'median', 'std', 'min', 'max', 'count']).T
```

```python
# Compute mode separately since it may return multiple values
mode = salesd[numerical_cols].mode().iloc[0]
stats_summary['mode'] = mode
```

```python
# Display results
print("Descriptive Statistics for Numerical Columns:")
print(stats_summary)
```

```python
# Interpretation (Example)
for col in numerical_cols:
    print(f"\nColumn: {col}")
    print(f"Mean: {stats_summary.loc[col, 'mean']:.2f} (Average value)")
    print(f"Median: {stats_summary.loc[col, 'median']:.2f} (Middle value)")
    print(f"Mode: {stats_summary.loc[col, 'mode']:.2f} (Most frequent value)")
    print(f"Std Dev: {stats_summary.loc[col, 'std']:.2f} (Spread of data)")
```

```python
#Data Visualization
import matplotlib.pyplot as plt
import seaborn as sns
# Histograms for numerical columns
```

```python
# Histograms for numerical columns
for col in numerical_cols:
    plt.figure(figsize=(8, 5))
    sns.histplot(salesd[col], bins=30, kde=True)
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Boxplots for numerical columns
for col in numerical_cols:
    plt.figure(figsize=(8, 5))
    sns.boxplot(y=salesd[col])
    plt.title(f'Boxplot of {col}')
    plt.ylabel(col)
    plt.show()
```

```python
# Bar charts for categorical columns
for col in numerical_cols:
    plt.figure(figsize=(10, 5))
    sns.countplot(x=salesd[col], order=salesd[col].value_counts().index)
    plt.title(f'Bar Chart of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# Standardization of Numerical Variables
scaler = StandardScaler()
data_scaled = salesd.copy()
data_scaled[numerical_cols] = scaler.fit_transform(salesd[numerical_cols])

print("\nStandardized Numerical Data:")
print(data_scaled[numerical_cols].head())
```

```python
# One-Hot Encoding for Categorical Variables
data_encoded = pd.get_dummies(data_scaled, columns=numerical_cols, drop_first=True)

print("\nDataset after One-Hot Encoding:")
print(data_encoded.head())
```
